## 1. Observational characteristics of interstellar clouds

**Q:** 

Describe briefly how to empirically determine...

* (a) ...the distance $D$ to a cloud.

* (c) ...the extinction $A_V$ of a cloud.

**A:**

a)

One way to measure the distance to clouds is by _spectroscopic parallax_. Spectroscopic parallax works by first taking a spectra of a star and determine its _spectral class_. Once we know this we can place the star in an HR-diagram and read of an expected absolute magnitude $M$ for such a star. Then we use photometry to measure the stars apparent magnitude $m$ and calculate the distance to the star from the _distance modulus_

$$
m-M = 5 \log(\frac{d}{10}) +A
$$

with the assumption that we know the extinction $A$. Here we of course assume that the stars in question are located in the cloud complex. This technique is not very precise, but the accuracy can be increased by measuring many stars in the cloud.

b)

In the context of giant molecular clouds we can picture the structure of these clouds as consisting of individual dark core clumps moving in random directions as a consequence of the gravitational field by the whole structure. The _virial velocity_---the velocity of a gas parcel that traverses the cloud in the free fall time $\tau_{ff}$---can then be used to calculate the mass of the whole structure. It is defined as

$$
V_{vir} = \left(\frac{GM}{R}\right)^{1/2}
$$

where we approximate the spacial structure as a sphere with radius $R$. It can be shown that the velocity dispersion $\Delta V$ in the cloud is roughly equal to $V_{vir}$. If we measure the velocity dispersion from spectral line broadening (typically CO-lines) and combine it with estimates of $R$ we can then calculate the mass $M$ of the cloud.

Another way to estimate the mass of clouds is by estimating the number densities in clouds. It can be done from the empirical relationships

$$
N_{CO}^{13} = 2.5 \cdot 10^{15} A_V 
$$

$$
N_H = 7.5 \cdot 10^5 N_{CO}
$$

where $N_{CO}^{13}$ is the column density of $^{13}C ^{16}O$ and $A_V$ is the extinction in the V-band. By estimating $A_V$ we can then estimate the column density of neutral hydrogen $N_H$. Once we have the column density we can derive the number density from spacial information of the cloud. The estimation of $A_V$ is explained in c).

c)

To determine the extinction of a cloud we imagine a cloud contained in a solid angle $\Omega_c$ and between the two distances $r_1$ and $r_2$ as seen in the figure below. If the cloud is not too opaque we may estimate $A_V$ from the obscuration of background stars. If the cloud itself is the only source of extinction, the stars that are at a distance $r > r_2$ will have a higher magnitude by $A_V$. From the distance modulus the distance to the stars increases with $m_V$ as

$$
\Delta \log r = 0.2 \Delta m_V.
$$

However this relation only holds for the unobscured stars in front of the cloud and if we make a plot of the apparent magnitude versus the number of stars, we see a shift from the expected linear curve as can be seen in the plot below. $A_V$ can then be directly determined from such a plot.

d)

The pressure of a cloud can be estimated from temperature measurements. If we for example assume an equation of state for the cloud as $P = \rho c_S^2$ and $c_S = \sqrt{\mathcal{R} T / \mu}$ where $P$ is the pressure, $c_S$ is the sound speed of the gas, $\rho$ is the density, $\mathcal{R}$ is the gas constant, $\mu$ is the mean molecular weight and $T$ is the temperature, then we can estimate the pressure from the temperature assuming we have knowledge about the density and mean molecular weight. The temperature can for example be estimated from CO-lines. 

e)

We can estimate the age of a cloud if we assume that all the stars in in the cloud formed at the same time and from the same cloud. Then they should all have similar properties except from their masses. If they have a range in masses we should see a large spread of stars along the main sequence. As the star ages, the higher mass stars burn out their hydrogen faster than the lower mass stars and we can therefore expect to see the main sequence of stars to become shorter over time. If you can find out exactly what type of star is just leaving the main sequence to become a red giant and you know from stellar models the expected age of such a star, then the age of the cloud should be roughly the same as that star since all stars formed simultaneously.

To solve the differential equation I use a ODE solver in scipy. 


```python
%matplotlib inline
from __future__ import division
from scipy import integrate

P_ext = 8*10**-12		# -13 in Pa, -12 in cgs units (bary 'barr-ee')
rho_ext = P_ext/(cs.value**2)		#7.7*10^-22 g cm^-3
rho_c = 7.5*10**-20
rho_ratio = rho_ext/rho_c	#0.1
psi_ext = -sp.log(P_ext/(rho_c*cs.value**2))


def f(y, t):
    y0 = y[0]
    y1 = y[1]

    f0 = y1
    f1 = sp.exp(-y0) - (2/t)*y1
    return [f0, f1]

y_init = [0, 0]         #create a list with initial conditions
t = sp.linspace(0.0001, 6, 5000)         #create a list with equidistant integration "steps" (start,stop, nr of steps)

#Python ODE solver. Takes the function f, inital conditions y_init and steps t as input
solu = sp.integrate.odeint(f, y_init, t)
psi = solu[:,0]          #saves the psi values in list psi
psi_diff = solu[:,1]     #saves the derivative of psi in list psi_diff


rho_frac = sp.exp(-psi)     #calculates the fraction rho/rho_c

xi_ext_index_list = sp.where(rho_frac < rho_ratio)     #selects the indexes from the list of rho/rho_c values that are < rho_ratio 
xi_ext_index = xi_ext_index_list[0][0]         #select first element in list.
xi_ext = t[xi_ext_index]         #uses the selected index to pick out the radius where the bonnor-evert sphere ends 

# plot the results
font = {'family' : 'ubuntu',
        'size'   : 13}

plt.rc('font', **font)
plt.figure(1)
plt.plot(t, psi, color='blue', linewidth = 2, label='$\psi$')
plt.plot(t, rho_frac, color='red', linewidth = 2,label='$\\rho$/$\\rho_c$')
plt.xlabel('Nondimentsional radius $\\xi$')
plt.legend()
plt.show()
```

This reproduces plot one perfectly. 

To produce the Second plot I create a for-loop that loops over a list with different external pressures from $10^{-10}$ to $10^{-15}$ Ba (barye). In each loop the differential equation is solved and the dimensionless mass $m$ is calculated from

$$

```python
m = np.array([])         #create an empty array for storing dimensionless mass values
rho_ext_frac = np.array([])         #create an empty array for storing values of the fraction rho_c/rho_ext
PextList = np.linspace(10**-10,10**-15,10000)         #make a list of external pressures

##########################
#
# this for-loop loopes over the list of external pressures creating a
# different endpoint on the bonnor-ebert sphere in each loop.
#
#########################
for P_ext in PextList:
    rho_ext = P_ext/(cs.value**2)
    rho_c = 7.5*10**-20
    rho_ratio = rho_ext/rho_c	#0.1
    psi_ext = -sp.log(P_ext/(rho_c*cs.value**2))

    def f(y, t):
        y0 = y[0]
        y1 = y[1]

        f0 = y1
        f1 = sp.exp(-y0) - (2/t)*y1
        return [f0, f1]

    
    
    y_init = [0, 0]         #create a list with initial conditions
    t = sp.linspace(0.0001, 500, 10000)         #create a list with equidistant integration "steps" (start,stop, nr of steps)

    #Python ODE solver. Takes the function f, inital conditions y_init and steps t as input
    solu = sp.integrate.odeint(f, y_init, t)
    psi = solu[:,0]          #saves the psi values in list psi
    psi_diff = solu[:,1]     #saves the derivative of psi in list psi_diff


    rho_frac = sp.exp(-psi)     #calculates the fraction rho/rho_c

    xi_ext_index_1ist = sp.where(rho_frac < rho_ratio)     #selects the indexes from the list of rho/rho_c values that are < rho_ratio 
    xi_ext_index = xi_ext_index_1ist[0][0]         #select first element in list.
    xi_ext = t[xi_ext_index]         #uses the selected index to pick out the radius where the bonnor-evert sphere ends 
    
    m_calc = np.sqrt(rho_ext/(4*np.pi*rho_c))*((xi_ext**2)*psi_diff[xi_ext_index])     #calculates the dimensionless mass
    rho_ext_frac = np.append(rho_ext_frac, np.log10((rho_c/rho_ext)))     #calculates rho_c/rho_ext and appends it to an array
    m = np.append(m,m_calc)     #append calculated dimensionless mass to array
    
    
    
# plot the results
font = {'family' : 'ubuntu',
        'weight' : 'bold',
        'size'   : 13}

plt.rc('font', **font)
plt.figure(2)
plt.plot(rho_ext_frac, m, color='blue', linewidth = 2)
plt.xlabel('Density Contrast $\log(\\rho_c/\\rho_0)$')
plt.ylabel('Cloud mass $m$')
plt.show()
```
