## 2. Dust-gas dynamics

**Q:**

The orbital velocity of gas in an axisymmetric circum-
stellar disk is not only dependent on the gravity from the central star, but also on the
gas pressure gradient. The midplane radial equation of motion for gas in cylindrical
coordinates is thus

$$
\ddot{r}-r \dot{\varphi}^2 = -\frac{GM_{\star}}{r^2} - \frac{1}{\rho}\frac{\partial P}{\partial r}
$$

where $r$ and $\varphi$ are the cylindrical radius and azimuthal angle respectively, $G = 6.67 \cdot
10^{−11}$ m$^3$ kg$^{−1}$ s$^{−2}$ is the gravitational constant, $M$ is the stellar mass, $P$ is the gas
pressure, and $ρ$ is the gas mass density. Assume that the gas is isothermal in circular
orbit around the star, and that the radial midplane density is $\rho(r) = \rho_0(r/r_0 )^{−\alpha}$ .

(a) What is the orbital speed $v_{gas}(r) = r\dot{\varphi}(r)$ of the gas as a function of radius $r$? Use the isothermal equation of state.

(b) A dust grain of size s is orbiting at Keplerian speed $v_{dust}(r)$ at radius $r$, and so
will experience a relative velocity with respect to the gas, giving rise to “Epstein drag"

$$
F_{drag} = -\frac{4}{3}\pi s^2 \rho v_{rel} v_{th}
$$

where $v_{rel} = v_{dust} - v_{gas}$ is the relative velocity between the dust and the gas, and

$$
v_{th} = \sqrt{ \frac{8 k_B T}{\pi \mu m_H} }
$$

is the thermal speed of the gas molecules, where $\mu$ is the mean molecular weight,
$k_B = 1.38 \cdot 10^{−23}$ J K$^{−1}$ is the Boltzmann constant, and $m_H = 1.66 \cdot 10^{−27}$ kg is
the hydrogen atom mass. Let the stopping time be defined by

$$
t_{stop} = \frac{m v_{rel}}{F_{drag}}
$$

where $m = 4\pi s^3 \rho_{dust} /3$ is the mass for a spherical dust grain of size s and density
\rho_{dust}. What is the stopping time as a function of the radius $r$ and dust grain size
$s$? What is the dimensionless stopping time $T_{stop} = \dot{\varphi} t_{stop} ?

(c) Compute the dimensionless stopping time $T_{stop}$ for the specific case $s = 2 \mu m,
r = 1$ AU, $T = 100$ K, $M_{\star} = 1 M_{\odot}$ , $r_0 = 1$ AU, $\alpha = 7/2$, $\rho_0 = 2 \cdot 10^{−12}$ kg cm$^{−3}$ ,
$\rho_{dust} = 3 \cdot 10^3$ kg cm$^{−3}$, and $\mu = 2$.



**A:**

Assuming an equation of state as

$$
P = \rho c_S^2
$$

where

$$
\rho = \rho_0 r^{-\alpha} r_0^{\alpha}
$$

and

$$
c_S = \sqrt{\frac{RT}{\mu}}
$$

In this final step I create a program that calculates $v_{gas}$, $t_{stop}$ and $T_{stop}$:

```python
#initialising values
G = con.G
R = con.R
pi = np.pi
mu = 2 * u.kilogram * u.mol**-1
mH = con.m_p
k = con.k_B
rhoDust = 3*10**3 *u.kilogram *u.meter**-3
rho0 = (2*10**-12 * u.kilogram * u.centimeter**-3).si
r0 = con.au
alpha = 7./2
r = con.au
T = 100 *u.Kelvin
Mstar = con.M_sun
s = 2*10**-6 *u.meter

#calculates gas speed
vGas = np.sqrt( ((G*Mstar)/r) - ((R*T*alpha) / mu))

#clculates stopping time
tStop = ( (rhoDust * s) / (rho0 * r**-alpha * r0**alpha) ) *np.sqrt( (pi * mu.value * mH) / (8*k * T) )

#calculates dimensionless stopping time
TStop = (1/r)* vGas * ( (rhoDust * s) / (rho0 * r**-alpha * r0**alpha) ) *np.sqrt( (pi * mu.value * mH) / (8*k * T) )
```

## 3. Gravitational instability

A criterion commonly used for determining disk
stability against gravity is the Toomre criterion that states that a disk annulus can
start to fragment when

$$
Q \equiv \frac{c_S \Omega}{\pi G \Sigma} < 1
$$

where $c_S$  is the sound speed, $\Omega$ is the orbital frequency, $G$ is the gravitational constant,
and $\Sigma$ is the surface mass density.

(a) Using reasonable assumptions, compute $Q$ as a function of radius for the MSMN.

(b) Compute the critical surface density as a function of radius (i.e. for $Q = 1$) and
compare it to the MSMN.

(c) A ‘clump’ in a disk will resist shear if it is Roche unstable, that is, if

$$
\rho > \rho_{Roche} \sim \frac{3.5 M_{\star}}{r^3}
$$

where $\rho$ is the volume mass density, $r$ is the radius and $M_{\star}$ is the mass of the central
star. Using some reasonable assumptions (e.g. for the scale height), compute the
Roche critical surface density $\rho$ Roche and compare it to the MSMN and Toomre
critical densities above.

a)

I assume isothermal equation of state with sound speed as

$$
c_S = \sqrt{\frac{R T}{\mu}}.
$$

I also assume that the angular frequency can be expressed in terms of the gas speed in <a href="#2.-Dust-gas-dynamics">problem 2</a>

$$
\Omega = \frac{v}{r} = \frac{1}{r} \sqrt{ \frac{G M_{\star} }{r} -\frac{R T \alpha}{\mu} }
$$

and I assume that the surface density can be described as the one derived in <a href="#1.-The-Minimum-Solar-Mass-Nebula-(MSMN)">problem 1</a>

$$
\Sigma(r) = 52.352\cdot r^{-1.751}.
$$

Finally I assume that $T = 100$ K, $\alpha = 7/2$, $\mu = 2$ and $M_{\star} = M_{\odot}$. With this I calculate $Q$ as a function of $r$ below:

```python
#initialising values
G = con.G
pi = np.pi
R = con.R
T = 100 *u.Kelvin
Mstar = con.M_sun
alpha = 7./2
cs = np.sqrt((R*T)/mu)
mu = 2 * u.kilogram * u.mol**-1
```

```python
r = np.linspace(0.00001*AU.value,2*AU.value, 10000) * u.meter
vGas = np.sqrt( ((G*Mstar)/r) - ((R*T*alpha) / mu))
sigma = np.e**constants[1] * r.value**constants[0]
Q = ( (cs*(vGas/r)) / (pi*G*sigma) )
plt.plot(r.value/con.au.value,Q.value ,linewidth=2, color='#FF7373')
plt.xlim(-0.0001,0.01)
plt.ylim(0,8)
plt.xlabel('Radius [AU]')
plt.ylabel('Q')
plt.grid()
```

From this plot we can see that with these assumptions the disk is only unstable very close to the sun, inside the radius of the sun. This means that the disk is not unstable at all.

b)

To derive the critical surface density as a function of $r$ I put the expression equal to one and solve for $\Sigma$

$$
Q = 1 = \frac{c_S \Omega}{\pi G \Sigma} \Leftrightarrow \Sigma = \frac{c_S \Omega}{\pi G}.
$$

I plot this expression below:

```python
r = np.linspace(0.05*AU.value,5*AU.value, 10000) * u.meter
vGas = np.sqrt( ((G*Mstar)/r) - ((R*T*alpha) / mu))
sigmaCrit = ((cs*(vGas/r)) / (pi*G))
plt.plot(r.value/con.au.value,sigmaCrit.value, linewidth=2, color='black', label='Critical')
plt.xlabel('Radius [AU]')
plt.ylabel('$\Sigma$_Critical [kg/m$^2$]')
plt.xlim(-0.05,2)
plt.ylim(0,100000)

y3=np.e**constants[1]*r**constants[0]
plt.plot(r.value/AU.value, y3.value, linewidth=2, color='#FF7373', label='MSMN')
plt.xlabel('Radius [AU]')
plt.ylabel('Surface density $\Sigma$ [kg m$^{-2}$]')
plt.grid()
plt.legend()
```

From this plot we see that the critical surface density is much higher than the predicted values from the MSMN-model.

c)

To convert the roche volume density into a surface density I multiply it with $h= 0.05r$, which is the criteria for the thin disk approximation and get

$$
\rho > \rho_{Roche} \cdot h \sim \frac{0.05 \cdot 3.5 M _{\odot} }{r^2}
$$

and then I plot this function together with the MSMN and critical density:

```python
r = np.linspace(0.15*AU.value,5*AU.value, 10000) * u.meter
h=cs/(vGas/r)
roche = (3.5*Mstar*h) / r**3
plt.figure(1)
plt.plot(r.value/AU.value,roche.value, linewidth=2, color='blue', label='Roche')
plt.xlabel('Radius [AU]')
plt.ylabel('$\\rho$_Roche [kg/m$^2$]')
plt.xlim(0.002,3)

plt.plot(r.value/con.au.value,sigmaCrit.value, linewidth=2, color='black', label='Critical')
plt.xlabel('Radius [AU]')
plt.ylabel('$\Sigma$_Critical [kg/m$^2$]')
plt.ylim(0,800000)

y3=np.e**constants[1]*r**constants[0]
plt.plot(r.value/AU.value, y3.value, linewidth=2, color='#FF7373', label='MSMN')
plt.xlabel('Radius [AU]')
plt.ylabel('Surface density $\Sigma$ [kg m$^{-2}$]')
plt.grid()
plt.legend()
```

Now we can see that the MSMN is much lower than both the critical density and the density at which the matter starts to resist shear in the disk.
