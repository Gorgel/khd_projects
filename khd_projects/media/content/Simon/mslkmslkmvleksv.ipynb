## Temperatures

When dealing with physics in general and with interstellar clouds in particular we will come accros differnet definitions of temperature. In this part we define and discuss some of these temperatures. We will look more closely at the following types:

* Excitation temperature, $T_{ex}$
* Gas temperature, $T_{gas}$
* Dust temerature, $T_{dust}$
* Antenna temperature, $T_B$

Other temperatures(Ex: rotation temp, nuclear spin temperature etc.) often fall in a subcategory of any these temperatures. All these temperatures can be the same, but in general they are not. Now we go through each one in the list, starting with the excitation temperature.

### Excitation temperature $T_{ex}$

For a population of particles where $n_i$ is the number of particles in an upper state, $n_j$ is the number of particles in a lower state, $g_i$ and $g_j$ are the statistical weights, then the **exitation temperature** $T_{ex}$ is defined from the boltzman factor as:


$$
\frac{n_i}{n_j} = \frac{g_i}{g_j} e^{-\frac{h\nu}{k_B T_{ex}}}
$$

Thus, if we find a system with a specific ratio of level populations, we then expect it to have a temperature of $T_{ex}$.

For line transitions we define $T_{ex}$ as the line emission intensity that matches the emission from a blackbody of temperature $T_{ex}$. Mathematically we express this as


$$
B_{\nu}(T_{ex}) = S_{\nu}(line) = \frac{2h \nu^3_{ij}}{c^2} \frac{1}{\frac{n_i g_j \phi(\nu)_{abs}}{n_j g_i \psi(\nu)_{em}}-1}
$$

where $\phi(\nu)_{abs}$ and $\psi(\nu)_{em}$ are the absorption and emission profiles. When we have "complete redistribution"---that is whenthe system is dominated by collisions---then $\phi(\nu)_{abs} = \psi(\nu)_{em}$. Below I plot a black body function

```python
from __future__ import division
from IPython.html.widgets import interact
%matplotlib inline

def boltFactor(Tex):
    nu =  2.47*10**15
    gi=2
    gj=4
    ninj = (gi/gj)*np.e**(-(con.h.value*nu)/(con.k_B.value*Tex))
    plt.figure(1)
    CS = plt.scatter(nu, ninj, color='red')
    font = {'family' : 'ubuntu',
        'weight' : 'bold',
        'size'   : 13}
    plt.rc('font', **font)
    plt.title('$n_i / n_j = $' + str(ninj))
    plt.xlabel('$nu$')
    plt.ylabel('$n_i / n_j$')
    plt.ylim(0,1)
    plt.show()
    

def planck_nu(T):
    nu = np.linspace(10,10**15.5,100)
    Bnu = ( (2*con.h*(nu*u.Hz)**3) / (con.c**2) ) * ( 1/ ( np.e**( (con.h*(nu*u.Hz)) / (con.k_B*T*u.Kelvin) ) -1 ))
    plt.plot(nu, Bnu.value,color='black', linewidth='2')
    plt.show()
    
def planck_lam(T):
    lam = np.linspace(0,10**-5.5,100)
    Blam = ( (2*con.h*con.c**2) / ((lam*u.meter)**5) ) * ( 1/ ( np.e**( (con.h*con.c) / ((lam*u.meter)*con.k_B*T*u.Kelvin) ) -1 ))
    plt.plot(lam, Blam.value, color='black', linewidth='2')
    plt.show()
```

For $T_{ex}$ to be well defined, it is not nessesary for the emitting medium to be in thermal equilibrium. In the general case the population of energy levels are obtained from the statistical equilibrium equations


$$
\frac{dn_j}{dt} = (\text{transition to j}) - (\text{transition from j}) = \sum_k \sum_i n_i R_{ij}^k - n_j \sum_k \sum_i R_{ji}^k
$$

where $k$ is different processes (e.g spontaneous emission, colissional excitation etc.) and $R_{ij}$ is the rate for a single particle in state $i$ to transit to state $j$by process $k$. In general some rates (e.g radiative excitation) depend on the radiation field, and therefore on the solution to the radiative transfer equation. Since teh radiative transfer also depends on the level population we have to solve radiative transfer and level populations simultaniously.

### Gas and dust temperatures $T_{gas}, T_{dust}$

Because interstellar clouds are often transparent to thermal radiation, the heating and cooliing processes for gas and dust can cause them to have different temperatures; $T_{gas} \neq T_{dust}$. The temperature is related to the energy density through the energy balance equation

$$
\frac{d\epsilon}{dt} = \sum_i \Gamma_i(n,T) - \sum_j \Lambda_j(n,T) - \nabla \cdot (\sigma \nabla T)
$$

where $\Gamma_i$ are heating terms (usually expressed in terms erg cm$^3$ s$^{-1}$), $\Lambda_j$ are cooling terms, and the last term is conduction with conductivity $\sigma$ wich e usually neglect in the ISM. In steady state, equilibrium is reached in a cooling time

$$
\tau_{cool} = \frac{\epsilon}{\sum_j \Lambda_j(n,T)}, \text{    with} \frac{d\epsilon}{dt} = 0
$$

Neglecting conduction, we equate heating with cooling

$$
\sum_i \Gamma_i(n,T) n= \sum_j\Gamma(n,T)
$$

Examples of important heating mechanisms in the ISM are photoelectric emission, cosmic rays, shocks, friction(gas-dust can also act as cooler). Cooling mechanisms are forbidden atomic transition line emission, molecular emission and adiabatic expansion.

#### Heating $\Gamma_i$

* Cosmic ray heating:
    $$
    \Gamma_{CR} = \sigma_0n_{Hz} \Delta E
    $$

    $\sigma_0 \Delta E$ is determined empirically to be $\sigma_0 \Delta E \sim 10^{-28}$ erg s$^{-1}$ molecule$^{-1}$
    
    
* Photoelectric heating, due to photons kicking out electrons from dust and thus heating the gas:

    $$
    \Gamma_{PE} = 4\pi n_{dust}\sigma_{dust} \underbrace{\epsilon_{PE}}_{\text{efficiency}} \int_{FUV} \underbrace{J_{nu}}_{\text{mean intensity}} d\nu
    $$
    
    If the only source of UV photons is the stellar background(the "Habing flux) then

    $$
    \Gamma_{PE} = 3\cdot 10^{-11}\left(\frac{n_H}{10^3 \text{cm}^{-3}}\right) \text{eV cm}^{-3} \text{s}^{-1} 
    $$


* Gas-dust collisions:

    $$
    \Gamma_{g-d} \propto (T_{dust}-T_{gas})
    $$
    
    If $T_{dust} > T_{gas}$, the gas is heated by the dust. If $T_{dust} < T_{gas}$ it is cooled

#### Cooling $\Gamma_i$

* Spectral line:
    
    $$
    \Lambda_{line} = h \nu_{ij} n_i C_{ij}
    $$
    
    where $C_{ij}$ is the collision rate coefficient from state i to j. The assumption is that all collisional transitions from i to j result in a radiative transition from j to i, with the energy being lost from the cloud through the photon(also assuming the cloud is opticaly thin!). Alternatively if the levels {$n_i$} are determined from statistical equilibrium, the cooling rate can be determined as
    
    $$
    \Lambda_{line} = h\nu_{ij} n_i A_{ij}
    $$
    
    wehre $A_{ij}$ is the Einstein coefficient for spontanious emission.
    
    
* Dust cooling: Dust radiates from its surface as a modified blackbody

    $$
    \Lambda_{dust} = \frac{1}{c} \oint \int \underbrace{\kappa_{\nu}}_{\text{Opacity}} B_{\nu}(T_{dust})d\nu d\Omega
    $$
    
    
* Expansion cooling: 

    $$
    \Lambda_{exp} = -\frac{Du}{Dt} = -\frac{P}{\rho} \frac{D\rho}{Dt} - u\nabla\cdot \mathbf{v} \approx k_BT \frac{dn(?)}{dt} = \underbrace{n k_B T \left[\frac{a}{v}+\frac{2v}{r}\right]}_{\text{Wind expansion}}
    $$
    
    where $ \frac{D}{Dt}= \frac{\delta}{\delta t} + \mathbf{v} \cdot \nabla$ is the "convective derivative", $u$ is the energy density, $P$ is the pressure, $\rho$ is the density, $v$ is the velocity and $a$ is the acceleration.
    

### Antenna temperature $T_B$

The antenna temperature is the temperature of a black body that would produce the measured intensity (averaged over the beam). In case (ii) the cloud fills the beam and $T_B = T_{cloud}$. In case (i), the beam is larger than the solid angle of the cloud, introducing "beam dilution"; the antenna temperature is then derived from

$$
B(T_{cloud}) \Delta \Omega_{cloud} = B(T_B) \Delta \Omega_{beam}
$$

## The Virial theorem

Why do not giant molecular clouds collapse onto themselves? what gives them support to counteract gravity? To approach the answer to this question we study the virial theorem. Recall the equation of hydrostatic equilibrium

$$
-\nabla \cdot P - \rho \nabla \Phi_g = 0
$$

where $P$ is the pressure, $\rho$ is the density and $\Phi_g$ is the gravitational potential. We now generalise this quation to allow for time-dependent moving fluid elements, in analogy with newtons law $F = ma$

$$
\rho \frac{D \mathbf{u}}{Dt} = -\nabla P -\rho \nabla \Phi_g + \frac{1}{c} \mathbf{j} \times \mathbf{B}
$$

where $\frac{D}{Dt}$ is the convective derivative, and $\frac{1}{c} \mathbf{j} \times \mathbf{B}$ is the lorentz force from a magnetic field $\mathbf{B}$ acting on a current density $\mathbf{j}$. These are related through Ampere's law

$$
\nabla \times \textbf{B} = \frac{4\pi}{c} \mathbf{j}
$$

so

$$
\frac{1}{c} (\mathbf{j} \times \mathbf{B} = \frac{1}{4 \pi} (\nabla \times \mathbf{B})\times \mathbf{B} = \frac{1}{4\pi}(\mathbf{B} \cdot \nabla) \mathbf{B} - \frac{1}{8\pi} \nabla \vert \mathbf{B}\vert^2
$$

where we have used the vector identity for cross product products

$$
\mathbf{A} \times ( \mathbf{B} \times \mathbf{C}) = (\mathbf{A} \cdot \mathbf{C})\mathbf{B} - (\mathbf{A} \cdot \mathbf{B})\mathbf{C}
$$

The First term expresses the tensition from the curvature of magnetic field lines, while the second is the gradient of the magnetic pressure $\frac{\vert \mathbf{B} \vert^2}{8\pi}$, resulting from the crowding of magnetic field lines.

Summarising, we have the equation

$$
\rho \frac{D \mathbf{u}}{Dt} = -\nabla P - \rho \nabla \Phi_g + \frac{1}{4\pi}(\mathbf{B} \cdot \nabla)\mathbf{B} - \frac{1}{8\pi} \nabla \vert \mathbf{B} \vert^2 
$$

which describes the local behaviour of a fluid under various forces. To study the *global* roperties of the fluid, we integrate the equation over the full volume. This is a bit tedious and is shown in Apendix D of S&P. The resulting equation has the same terms but in integrated form:

$$
\frac{1}{2} \frac{\delta^2 I}{\delta t^2} = 2T+ 2U+W +M
$$

where

\begin{align}
I \equiv \int \rho \vert \mathbf{r} \vert^2 dV \phantom{abcd} \text{is the moment of inertia} \\
T = \frac{1}{2} \int \rho \vert \mathbf{u} \vert^2 dV \phantom{abcd} \text{is the total kinetic enrgy} \\
U = \frac{3}{2} \int n k_B T dV \phantom{abcd} \text{is the total thermal energy}\\
W = \frac{1}{2} \int \rho \Phi_g dV \phantom{abcd} \text{is the gravitational potential energy}\\
M =  \frac{1}{8\pi} \int \vert \mathbf{B} \vert^2 dV \phantom{abcd} \text{is the magnetic energy}
\end{align}

**Note: Only W can be negative!**


If $ \vert 2T +2U + M \vert < \vert W \vert$ we have a gravitationall collapse. Assuming $\vert W \vert$ is completely dominating, we have free fall. With $M$ being the mass of the cloud and $R$ its characteristic radius, $W \approx \frac{-GM^2}{R}$. Approximating $I\approx MR^2$ then

$$
\frac{1}{2} \frac{\delta^2 I}{\delta t^2} = W \Rightarrow \frac{1}{2} \frac{\delta^2 ( MR^2)}{\delta t^2} = \frac{-GM^2}{R}
$$

or

$$
\frac{\delta^2 R^2}{\delta t^2} = \frac{2GM}{R}
$$

and with some sloppy handwaveing we have

$$
\Rightarrow \frac{\delta^2 R^2}{\delta t^2} \approx \frac{R^2}{\tau^2_{ff}} = \frac{2GM}{R} \Rightarrow \tau_{ff} = \sqrt{\frac{R^3}{2GM}} = 7 \cdot 10^6 \phantom{a} \text{yr} \left(\frac{M}{10^5 M_{\odot}} \right)^{-1/2} \left(\frac{R}{25\text{pc}} \right)^{3/2}
$$

This hints that a GMC (Giant molecular cloud) collapse within a few Myr, while they are observed to alst longer. The condition for long-term stability is

$$
\frac{\delta^2 I}{\delta t^2} \approx 0
$$

wich results in the **Virial Theorem**:

In stars, the termal pressure balances gravity. Could this be the case for GMCs? If we focus on the term

$$
U = \frac{3}{2} \int P dV = \frac{3}{2} \int n k_B T dV \approx \frac{M \Re T}{\mu}
$$

where $\Re$ is the gas constant, then

$$
\left \vert \frac{U}{W} \right \vert \approx \frac{M \Re T}{\mu} \left( \frac{GM^2}{R} \right)^{-1} = 3 \cdot 10^{-3} \left(\frac{M}{10^5 M_{\odot}} \right)^{-1} \left( \frac{R}{\text{25pc}}\right) \left( \frac{T}{15\text{K}} \right)
$$

Thus, for typical GMC values, $ \vert U \vert << \vert W \vert$ and cannot be responsible for te support. What about magnetic support?

$$
M \approx \frac{\vert \mathbf{B}^2 \vert R^3}{6\pi}
$$

so

$$
\left \vert \frac{M}{W} \right \vert \approx \frac{\vert \mathbf{B}^2 \vert R^3}{6\pi} \left( \frac{GM^2}{R} \right)^{-1} = 0.3 \left(\frac{B}{20 \mu \text{G}} \right)^2 \left( \frac{R}{25\text{pc}} \right)^4 \left( \frac{M}{10^5 M_{\odot}} \right)^{-2}
$$

Quite close! But magnetic pressure has the peculiarity of only working across the field lines, thus being highly anisotropic. Along the field lines the cloud in principle stil collapse, forming a huge sheet. A solution might be that the magnetic field is not well ordered on large scales, so that there is a no preffered direction. These distortions of a global magentic field can arise from MHD waves, something we will get back to later. Finally, let us check the last term T, corresponding to bulk motion of the cloud

$$
T \approx \frac{1}{2}M \Delta v^2
$$

so

$$
\frac{T}{\vert W \vert} = 0.5 \left( \frac{\Delta v}{4 \text{km} \text{s}^{-1}} \right)^2 \left(\frac{M}{10^5 M_{\odot}} \right)^{-1} \left(\frac{R}{25\text{pc}} \right)
$$

(4km/s is a typical dispersion in GMCs)
